## 3. Ejecutar una query de lectura

El binding expone un `QueryResult` iterable cuya forma exacta depende de la versión instalada. El helper `_unwrap_query` en `shared.viz` cubre ambas APIs (v0.4.16 directa vs v0.4.20+ con `.kind`).

```python
from shared.viz import _unwrap_query

res = graph.execute_nql("find n.x, n.label from (n:Smoke)")
qr = _unwrap_query(res)
print("columns:", list(qr.columns))
print("rows:")
for row in qr:
    print("  ", row)
```
